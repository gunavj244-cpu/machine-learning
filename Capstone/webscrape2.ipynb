```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

# ‚úÖ Setup Chrome options
options = Options()
options.add_argument('--headless')  # Run in background
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# ‚úÖ Flipkart search URL
url="https://www.flipkart.com/search?q=sony&otracker=AS_Query_HistoryAutoSuggest_2_0&otracker1=AS_Query_HistoryAutoSuggest_2_0&marketplace=FLIPKART&as-show=on&as=off&as-pos=2&as-type=HISTORY"
driver.get(url)

wait = WebDriverWait(driver, 15)
page = 1
max_pages = 50  # Adjust as needed

# ‚úÖ Lists to store data
product_names, product_prices, product_ratings, product_mrp, product_discount = [], [], [], [], []

while page <= max_pages:
    time.sleep(3)

    # Scroll to load all products
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

    # ‚úÖ Parse the HTML
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # ‚úÖ Updated container for product blocks
    products = soup.find_all("div", class_="yKfJKb row")  # latest Flipkart product container
    print(f"‚úÖ Page {page}: Found {len(products)} products")

    if len(products) == 0:
        print("üö´ No products found on this page. Stopping.")
        break

    for p in products:
        # Product name
        name = p.find("div", class_="KzDlHZ")
        product_names.append(name.get_text(strip=True) if name else "N/A")

        # Price
        price = p.find("div", class_="Nx9bqj _4b5DiR")
        product_prices.append(price.get_text(strip=True) if price else "N/A")

        # Rating
        rating = p.find("div", class_="XQDdHH")
        product_ratings.append(rating.get_text(strip=True) if rating else "N/A")

        # MRP
        mrp = p.find("div", class_="yRaY8j ZYYwLA")
        product_mrp.append(mrp.get_text(strip=True) if mrp else "N/A")

        # Discount
        discount = p.find("div", class_="UkUFwK")
        product_discount.append(discount.get_text(strip=True) if discount else "N/A")

    # ‚úÖ Move to next page
    try:
        nxt_button = wait.until(EC.element_to_be_clickable((By.XPATH, '//span[text()="Next"]')))
        driver.execute_script("arguments[0].scrollIntoView(true);", nxt_button)
        driver.execute_script("arguments[0].click();", nxt_button)
        print(f"‚û°Ô∏è Moving to next page ({page + 1})...\n")
        page += 1
        time.sleep(4)
    except Exception:
        print("üö´ No more pages found. Stopping.")
        break

# ‚úÖ Close browser
driver.quit()

# ‚úÖ Create and save DataFrame
data = {
    "Product_name": product_names,
    "Price": product_prices,
    "Rating": product_ratings,
    "MRP": product_mrp,
    "Discount": product_discount
}

df = pd.DataFrame(data)
df.drop_duplicates(inplace=True)

print(df.head())
print("\n‚úÖ Total rows:", len(df))

df.to_csv("flipkart_electronics.csv", index=False)
print("üìÅ Data saved as 'flipkart_electronics.csv'")
```
